Test korelasi: tentukan apakah setiap kolom non-target berkorelasi dengan kolom target, dengan menggambarkan juga scatter plot nya. Gunakan correlation test.
Tes Korelasi yang digunakan adalah Pearson's Correlation Coefficient, mengukur hubungan linier antar variabel. Secara matematis jika (σXY) adalah kovarian antara X dan Y, dan (σX) adalah simpangan baku dari X, maka koefisien korelasi Pearson ρ adalah

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np

col_names = ['id', 'pH', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity', 'OrganicCarbon', 'Trihalomethanes', 'Turbidity', 'Potability']
df = pd.read_csv('water_potability.csv', names=col_names)
```

```python
y = df['Potability']
```

```python
for i in range(1, 10):
    x1 = df[col_names[i]]
    corr1 = np.corrcoef(x1, y)
    print("Korelasi antara {} dan Potability: {}".format(col_names[i], corr1[0, 1]))
    plt.scatter(x1, y)
    plt.xlabel(col_names[i])
    plt.ylabel('Potability')
    plt.show()
```
